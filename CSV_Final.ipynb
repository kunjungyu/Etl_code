```python
# Standard
import glob
import os
import datetime
import pprint

# Third-Party
from bs4 import BeautifulSoup
import pandas as pd

def to_timestamp_and_time(time):   
    date = datetime.datetime.strptime(time,"%Y-%m-%d")
    date_str = datetime.datetime.strftime(date,"%Y/%m/%d")
    ts = date.timestamp()
    return ts,date_str

def to_timestamp_and_time_minsec(time):   
    date = datetime.datetime.strptime(time,"%Y-%m-%d %H:%M")
    date_str=datetime.datetime.strftime(date,"%Y/%m/%d %H:%M")
    ts = date.timestamp()
    return ts,date_str

def extracting_news(html):    

    with open(html, 'r', encoding = 'utf8') as f:
        soup = BeautifulSoup(f.read(), 'html5lib')
        
    #topic
    try:
        topic = soup.select_one('div.whitecon > h1').text.strip()
    except:
        topic = soup.select_one('div.conbox > h2').text.strip()
    
    #time
    my_datetime = soup.select_one('div.text > span').text
    try:
        timestamp,date_str = to_timestamp_and_time_minsec(my_datetime)
    except:
        timestamp,date_str = to_timestamp_and_time(my_datetime)
     
    
    #content
    content = ''
    for p in soup.select('div.text > p'):
        content += p.text
   
    return {'topic': topic, 'timestamp': timestamp, 'date_str': date_str, 'content': content}

if __name__ == '__main__':

    
    
    # Test 1
    #result = to_timestamp_and_time('2017-10-06')
    #print(result)
    
    # Test 2
    #result = to_timestamp_and_time('2017-10-06')
    
    # Test 3 
#     result = extracting_news('D:/Project_ETL/LN_Original/001TPK/642999.html')
#     pprint.pprint(result)
    

    columns = ['topic', 'timestamp', 'date_str', 'content']
    path = 'D:/Project_ETL/LN_Original/006可成/*'
    csv_file = 'D:/Project_ETL/LN_CSV/news006.csv' 

    total_row_data = []
    for raw_html in glob.glob(path):
        print(raw_html)
        # record is dictionary
        record = extracting_news(raw_html)
        
        df = pd.DataFrame.from_dict([record])
        #df.to_csv(epath,encoding='utf-8')
        
        if os.path.isfile(csv_file):
            df.to_csv(csv_file, columns=columns, index=False, header=False, mode='a')
        else:
            df.to_csv(csv_file, columns=columns, index=False)

# 6/7/8/9/10/11
```
